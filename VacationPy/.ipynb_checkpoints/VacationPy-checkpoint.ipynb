```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
from pprint import pprint
# Import API key
from config import g_key

weather_data = pd.read_csv(r'../output_data/cities.csv')
del weather_data['Unnamed: 0']
weather_data
```

```python
locations = weather_data[["latitude", "longitude"]]
humidity = weather_data['humidity']
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
ideal_weather = weather_data.loc[weather_data['humidity'] <80 ]
ideal_weather = ideal_weather.loc[ideal_weather['temperature']< 95]
ideal_weather = ideal_weather.loc[ideal_weather['temperature'] > 80]
```

```python


# geocoordinates
target_coordinates = weather_data[["latitude", "longitude"]]
target_search = 'hotel'
target_radius = 5000
target_type = "hotel"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"


# run a request using our params dictionary
response = requests.get(base_url, params=params)

hotel_list = response.json()

print(json.dumps(hotel_list, indent=4, sort_keys=True))
#hotel_df = response
#ideal_weather['hotel'] = hotel_df
```
