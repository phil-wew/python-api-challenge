```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
from pprint import pprint
# Import API key
from config import g_key

weather_data = pd.read_csv(r'../output_data/cities.csv')
del weather_data['Unnamed: 0']
weather_data
```

```python
locations = weather_data[["latitude", "longitude"]]
humidity = weather_data['humidity']
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
ideal_weather = weather_data.loc[(weather_data['humidity'] <80) & (weather_data['temperature']< 94) & (weather_data['temperature'] > 75), :]


ideal_weather.reset_index(inplace=True, drop = True)

ideal_weather = ideal_weather.dropna()

ideal_weather
```

```python
hotelList = []

# geocoordinates
target_search = ['hotels','motel','lodging','hostel','hotel']
target_radius = 5000
target_type = ['hotels','motel','lodging','hostel','hotel']



# set up a parameters dictionary
params = {
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}


for row in ideal_weather.index:
    # get lat, lng from df
    lat, lng = ideal_weather.loc[row, ["latitude", "longitude"]]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"

    # Use the search term: "International Airport" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # make request and print url
    result_list = requests.get(base_url, params=params).json()

    print(json.dumps(result_list, indent=4, sort_keys=True))

    try:
        hotelList.append(result_list["results"][0]["name"])

    except (KeyError, IndexError):
        hotelList.append(np.nan)

```

```python
ideal_weather.loc[:,'hotel'] = hotelList
```

```python
for hotel, city, country in zip(ideal_weather.hotel, ideal_weather.city, ideal_weather.country):
    print(hotel, city, country)
```

```python

# Using the template add the hotel marks to the heatmap
info_box_template = f"""
<dl>
<dt>Name</dt><dd>{hotel}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Country</dt><dd>{country}</dd>
</dl>
"""

dat = [info_box_template.format(hotel, city, country) for hotel, city, country in zip(ideal_weather.hotel, ideal_weather.city, ideal_weather.country)]
locations = ideal_weather.loc[:, ["latitude", "longitude"]]

humidity = ideal_weather.humidity
```

```python
# Plot Heatmap
fig = gmaps.figure()


info_layer = gmaps.symbol_layer(
            locations, info_box_content=dat)

fig.add_layer(info_layer)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)



# Assign the marker layer to a variable
markers = gmaps.marker_layer(locations)


# Add the layer to the map

fig.add_layer(markers)
fig
```
